```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("student_data", engine='openpyxl')
df.head()
```

```python
df['pass'] = df['final_marks'].apply(lambda x: 1 if x >= 10 else 0)
display(df.head())
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='failures', y='final_marks', data=df)
plt.title("Past Failures vs Final Marks")
plt.show()
```

```python
sns.boxplot(x='sex', y='final_marks', data=df)
plt.title("Gender vs Final Marks")
plt.show()
```

```python
sns.scatterplot(x='absences', y='final_marks', data=df)
plt.title("Absences vs Final Marks")
plt.show()
```

```python
plt.scatter(df['studytime'], df['final_marks'])
plt.xlabel("Study Time")
plt.ylabel("Final Marks")
plt.title("Study Time vs Final Marks")
plt.show()
```

```python
import zipfile
import os

# Define the path to the zipped file and the extraction directory
zip_file_path = "student_data"
extract_dir = "extracted_data"

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"File '{zip_file_path}' unzipped to '{extract_dir}'")
```

```python
import pandas as pd

# Assuming the CSV file inside the zip is named 'student_data.csv'
# Adjust the csv_file_name if the extracted CSV has a different name
csv_file_name = "student_data.csv"
csv_file_path = os.path.join(extract_dir, csv_file_name)

# Now, try to read the extracted CSV file with the correct separator
# Given previous issues, we'll keep encoding='latin1' and on_bad_lines='skip'
try:
    df = pd.read_csv(csv_file_path, sep=',', encoding='latin1', on_bad_lines='skip')
except UnicodeDecodeError:
    print("UnicodeDecodeError: Trying with default UTF-8 encoding.")
    df = pd.read_csv(csv_file_path, sep=',', on_bad_lines='skip')

display(df.head())
```
